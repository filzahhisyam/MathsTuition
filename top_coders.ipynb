# Literals

```python
# I want to ask sequence of questions to the computer and
# I want the computer to give me the answers.

# 1) Where are you going ? (and I want the computer to say Shopping)
# you can use a built in function in python called print
print("Shopping") # you have to create the statement

# 2) What are you going to buy ? (and I want the computer to say Television)
print("Television")

# 3) How many ? (and I want the computer to say 10)
print(10)

# 4) What is the price of the Television ? (and I want the computer to say 1455.55)
print(1455.55)

# 5) Is the Television available in the shop ? (and I want the computer to say yes)
# Technically in computers the answers like yes/no are called boolean values
print(True)
```

In computers the values like
<ol>
<li>Shopping</li>
<li>Television</li>
<li>10</li>
<li>1455.55</li>
<li>True</li>
</ol>
are called <b>Literals</b>

There are 4 types of literals and they are
<ol>
<li>String ("Shopping", "Television") - Values created using alphbets</li>
<li>Integer (10) - numbers without decimals</li>
<li>Float (1455.55) - numbers with decimals</li>
<li>Boolean (True / False)</li>
</ol>

Note: Always remember computers never go to school. Every single thing we must make it very clear.

1) To differentiate Strings (alphabet) from Numbers and Booleans we enclose the String literals with double quote.

2) Possible values for boolean literals is just 2 only. True and False. These 2 words are registered as keywords in python. Eventhough they also created by alphabets, we dont use double quote to enclose it.

3) Number literals (Integer, float) are not enclosed by double quote.

```python
# What is the total price to be paid (and I want computer to multiply quantity with price)
print(10 * 1455.55)
# Note: In previous statements we hardcode values like "Shopping", "Television", 10, 1455.55, True
# However now, we are not calculating and passing the total amount to be paid to the print function
# We are using the computer to calculate the total amount to be paid and tell us
```

```python
# You can pass more than 1 argument to the print function

print("Where are you going ?", "Shopping")
print("What are you buying ?", "Television")
print("How many are you going to buy ?", 10)
print("Do you know the price ?", 1455.55)
print("Is it available in the shop ?", True)
```

### Summary about print function

In python we have a built in function called <b>print</b>.

Print function, prints the literals/values in the monitor. The function print is used for output.

In python to call a built in function, you have to mention the <b>function name</b> followed by <b>( )</b>

Example: <b>print( )</b>

Between the ( ) you can pass values and we call that as <b>arguments</b>.
<pre>
print("Shopping")
print("Television")
print(10)
print(1455.55)
print(True)
</pre>
Shopping, Television, 10, 1455.55, True are arguments to the print function.

The print function can take more than 1 argument.

Example: <br/>
<b>print("Product:", "Television")</b><br/>
<b>print("Quantity:", 10)</b><br/>

Note: In the above example we are passing more than 1 argument. The arguments are seperated by comma.

We already know print function is used to output literals. Print function can also be used to output the value of an expression.

Example: <b>print(10 * 1455.55)</b>
